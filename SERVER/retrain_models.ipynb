```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
import joblib
import numpy as np

print("--- Loading Datasets ---")
# Make sure these CSV files are in the SERVER folder
final_df = pd.read_csv('odisha_final_data.csv')
soil_df = pd.read_csv('data_core.csv')
soil_df.columns = soil_df.columns.str.strip()

print("Data preparation complete.")

# --------------------------------------------------------------------------

# Cell 2: Train and Save the Yield Prediction Model
print("--- Training Yield Prediction Model ---")

# Feature Engineering for Yield Model
final_df_encoded = pd.get_dummies(final_df, columns=['District', 'Crop', 'Season'], drop_first=True)
X_yield = final_df_encoded.drop(columns=['State', 'Year', 'Production', 'Yield', 'Area Units', 'Production Units', 'Crop_Category'])
y_yield = final_df_encoded['Yield']

# Train/Test Split
X_train_yield, _, y_train_yield, _ = train_test_split(X_yield, y_yield, test_size=0.2, random_state=42)

# Train the Model
yield_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
yield_model.fit(X_train_yield, y_train_yield)

# Save the compatible Yield Model files
joblib.dump(yield_model, 'crop_yield_model.joblib', compress=3)
joblib.dump(X_yield.columns, 'model_columns.joblib')

print("✅ Yield Model trained and saved successfully.")

# --------------------------------------------------------------------------

# Cell 3: Train and Save the Fertilizer Recommendation Model
print("--- Training Fertilizer Recommendation Model ---")

# Prepare the data for the classifier
X_fert = soil_df[['Nitrogen', 'Potassium', 'Phosphorous', 'Crop Type']]
y_fert = soil_df['Fertilizer Name']

# One-hot encode the 'Crop Type'
X_fert_encoded = pd.get_dummies(X_fert, columns=['Crop Type'], drop_first=True)

# Train the classifier
X_train_fert, _, y_train_fert, _ = train_test_split(X_fert_encoded, y_fert, test_size=0.2, random_state=42)
fertilizer_model = RandomForestClassifier(n_estimators=100, random_state=42)
fertilizer_model.fit(X_train_fert, y_train_fert)

# Save the compatible Fertilizer Model files
joblib.dump(fertilizer_model, 'fertilizer_recommender_model.joblib', compress=3)
joblib.dump(X_fert_encoded.columns, 'fertilizer_model_columns.joblib')

print("✅ Fertilizer Model trained and saved successfully.")
print("\n--- All models are now compatible with Python 3.13. You can now run the server. ---")
```
